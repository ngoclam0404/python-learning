```python
import numpy as np

#from google.colab import drive
#drive.mount('/content/drive')

np.random.seed(42)
# Tạo ma trận điểm 5 học sinh × 4 môn (số nguyên 5-10)
diem = np.random.randint(5, 11, size=(5, 4))
print(diem)
'''(a) Tính điểm TB của từng môn (axis=0)
(b) Tính điểm TB của từng học sinh (axis=1)
(c) Chuẩn hóa: trừ điểm TB mỗi môn khỏi toàn bộ cột đó — 1 dòng, không vòng lặp
(d) Kiểm tra: np.mean(diem_chuan_hoa, axis=0) có ≈ 0 không?'''

#a
diem_tb_mon = np.mean(diem, axis = 0)
print(diem_tb_mon)
#b
diem_tb_hs = np.mean(diem, axis = 1)
print(diem_tb_hs)
#c
diem_chuan_hoa = diem - diem_tb_mon
print(diem_chuan_hoa)
#d
print(np.mean(diem_chuan_hoa, axis = 0))
```

```python
import numpy as np


diem      = np.load("/content/drive/MyDrive/python-learning/data/diem.npy")
thu_nhap  = np.load("/content/drive/MyDrive/python-learning/data/thu_nhap.npy")
du_no     = np.load("/content/drive/MyDrive/python-learning/data/du_no.npy")
nhom_no   = np.load("/content/drive/MyDrive/python-learning/data/nhom_no.npy")

print(diem.shape, thu_nhap.shape, du_no.shape, nhom_no.shape)

# Ghép thành ma trận (500, 3)
data = np.column_stack([diem, thu_nhap, du_no])
print(data.shape)   # phải ra (500, 3)

'''(a) Min-Max Normalization: (x - min) / (max - min)

Tính min và max theo từng cột — 1 dòng mỗi cái
Áp dụng công thức bằng broadcasting
Kiểm tra: min của mỗi cột chuẩn hóa = 0, max = 1

(b) Z-score Normalization: (x - mean) / std

Tương tự, dùng broadcasting
Kiểm tra: mean ≈ 0, std ≈ 1

(c) In bảng so sánh trước và sau chuẩn hóa — ví dụ 3 dòng đầu của mỗi cách.'''

#a
min_data = np.min(data, axis = 0)
max_data = np.max(data, axis = 0)

print(min_data)
print(max_data)

minmax = (data - min_data) / (max_data - min_data)
print(minmax)

#b
mean_data = np.mean(data, axis = 0)
std_data = np.std(data, axis = 0)

z_score = (data-mean_data)/std_data
print(z_score)

# Kiểm tra Min-Max
print("=== Min-Max ===")
print("Min mỗi cột:", np.min(minmax, axis=0))   # phải = [0, 0, 0]
print("Max mỗi cột:", np.max(minmax, axis=0))   # phải = [1, 1, 1]

# Kiểm tra Z-score
print("=== Z-score ===")
print("Mean mỗi cột:", np.mean(z_score, axis=0))  # phải ≈ [0, 0, 0]
print("Std mỗi cột: ", np.std(z_score, axis=0))   # phải ≈ [1, 1, 1]
```

```python
data = np.column_stack([diem, thu_nhap, du_no])
# Tạo mask cho 2 nhóm
du_dk  = diem >= 650   # đủ điều kiện
khong_dk = diem < 650  # không đủ điều kiện

print(f"Đủ ĐK:    {np.sum(du_dk)} khách")
print(f"Không ĐK: {np.sum(khong_dk)} khách")


#(a) Với mỗi nhóm, tính mean và std của cả 3 chỉ số (điểm, thu nhập, dư nợ)
#Nhóm đủ điều kiện
mean_du   = np.mean(data[du_dk], axis=0)   # shape (3,)
std_du   = np.std(data[du_dk], axis=0)


#Nhóm không đủ điều kiện
mean_khong = np.mean(data[khong_dk], axis=0)
std_khong   = np.std(data[khong_dk], axis=0)  # shape (3,)

#(b) In bảng
print(f"            Đủ điều kiện            Không đủ điều kiện")
print(f"Điểm TB:   {mean_du[0]}                    {mean_khong[0]}")
print(f"TN TB:     {mean_du[1]}                    {mean_khong[1]}")
print(f"DN TB:     {mean_du[2]}                    {mean_khong[2]}")


#(c)
print(f"Chênh lệch dư nợ TB: {mean_du[2]-mean_khong[2]}")
```

```python
'''(a) Tính Z-score cho từng cột của data — dùng lại z_score từ bài 1 hoặc tính lại
(b) Tạo mask outlier: hàng nào có ít nhất 1 cột có |z| > 3
(c) Đếm số hàng là outlier
(d) Lọc bỏ các hàng outlier — in shape của data còn lại'''

mean_data = np.mean(data, axis = 0)
std_data = np.std(data, axis = 0)

z_score = (data-mean_data)/std_data
print(z_score)

la_outlier = np.abs(z_score) > 3
mask = np.any(la_outlier, axis=1)

print(np.where(mask))
print(data[mask])
print(f"Số outlier:       {np.sum(mask)}")
print(f"Data còn lại:     {data[~mask].shape}")
```
