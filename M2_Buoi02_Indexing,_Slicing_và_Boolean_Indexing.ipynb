```python
import numpy as np
np.random.seed(42)
n = 500
diem     = np.random.normal(680, 80, n).clip(300, 850)
thu_nhap = np.random.lognormal(3, 0.5, n) * 1e6
du_no    = np.random.uniform(0, 500e6, n)
nhom_no  = np.random.choice([1,2,3,4,5], n, p=[0.5,0.25,0.15,0.07,0.03])

'''Bài 1 — Phân tích danh mục vay bằng boolean indexing
Không dùng vòng lặp Python:
(a) Đếm số khách ở từng nhóm nợ (1, 2, 3, 4, 5)
(b) Lấy chỉ số (index) của khách đủ điều kiện vay:
diem >= 650 VÀ nhom_no <= 2 VÀ du_no < 200e6
(c) Tính DTI (tỷ lệ nợ/thu nhập hàng năm) cho toàn bộ:
DTI = du_no / (thu_nhap * 12)
(d) Tìm chỉ số của khách có DTI > 5 (rủi ro cao)
(e) Lấy điểm tín dụng của nhóm rủi ro cao đó và tính điểm TB'''

#(a) Đếm số khách ở từng nhóm nợ (1, 2, 3, 4, 5)
print(len(nhom_no[nhom_no == 1]))
print(len(nhom_no[nhom_no == 2]))
print(len(nhom_no[nhom_no == 3]))
print(len(nhom_no[nhom_no == 4]))
print(len(nhom_no[nhom_no == 5]))

#(b) Lấy chỉ số (index) của khách đủ điều kiện vay: diem >= 650 VÀ nhom_no <= 2 VÀ du_no < 200e6
kh_idx = (diem >= 650) & (nhom_no <= 2) & (du_no < 200e6)

print(np.where(kh_idx)[0])

#(c) Tính DTI (tỷ lệ nợ/thu nhập hàng năm) cho toàn bộ: DTI = du_no / (thu_nhap * 12)
print(du_no / (thu_nhap * 12))

#(d) Tìm chỉ số của khách có DTI > 5 (rủi ro cao)
print(np.where(du_no / (thu_nhap * 12) > 5))

#(e) Lấy điểm tín dụng của nhóm rủi ro cao đó và tính điểm TB
print(np.mean(diem[du_no / (thu_nhap * 12) > 5]))
```

```python
'''Áp dụng quy tắc điều chỉnh điểm — không dùng vòng lặp:

nhom_no == 1 → cộng 20 điểm
nhom_no == 2 → giữ nguyên
nhom_no >= 3 → trừ 50 điểm

Sau đó clip về [300, 850]. In điểm TB trước và sau.'''

import numpy as np
np.random.seed(42)
n = 500
diem     = np.random.normal(680, 80, n).clip(300, 850)
thu_nhap = np.random.lognormal(3, 0.5, n) * 1e6
du_no    = np.random.uniform(0, 500e6, n)
nhom_no  = np.random.choice([1,2,3,4,5], n, p=[0.5,0.25,0.15,0.07,0.03])

diem_moi = diem.copy()
diem_moi[nhom_no == 1] += 20
diem_moi[nhom_no >= 3] -= 50
diem_moi = diem_moi.clip(300, 850)
print(np.mean(diem_moi))
print(np.mean(diem))
print(np.mean(diem_moi) - np.mean(diem))


# Trước khi sửa bug — tìm xem có giá trị nào vượt [300,850] không
print(diem_moi.min())   # có < 300 không?
print(diem_moi.max())   # có > 850 không?
print(f"số lượng KH có nhóm nợ 1: {len(diem_moi[nhom_no == 1])}, tổng số điểm tăng thêm: {len(diem_moi[nhom_no == 1]) * 20}")


print(f"số lượng KH có nhóm nợ >=3: {len(diem_moi[nhom_no >= 3])}, tổng số điểm giảm đi: {len(diem_moi[nhom_no >= 3])*50}")
```
