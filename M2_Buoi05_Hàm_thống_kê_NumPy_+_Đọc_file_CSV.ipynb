```python
import numpy as np
data = np.column_stack([diem, thu_nhap, du_no])

# Bước b — Thống kê mô tả đầy đủ cho mỗi cột
mean_data = np.mean(data, axis = 0)
median_data = np.median(data, axis = 0)
std_data = np.std(data, axis = 0)
min_data = np.min(data, axis = 0)
max_data = np.max(data, axis = 0)

# Bước c — Đếm NaN ở từng cột (không phải tổng)
sum_nan = np.sum(np.isnan(data), axis=0)

# Bước d — So sánh mean vs median từng cột
dif_mean_median = mean_data - median_data


# Bước e — Phát hiện outlier theo IQR cho TỪNG cột riêng, Đếm số outlier mỗi cột
q1 = np.percentile(data, 25, axis = 0)
q3 = np.percentile(data, 75, axis = 0)
iqr = q3- q1
lower = np.maximum(q1 - 1.5 * iqr,0)
upper = q3 + 1.5 * iqr

outlier = [lower, upper]

so_outlier =  np.sum(((data < lower) | (data > upper)), axis=0)
# Bước f — In báo cáo tổng hợp
print("=" * 60)
print(f"{'CHỈ SỐ':<15} {'DIEM':>10} {'THU_NHAP':>12} {'DU_NO':>12}")
print("=" * 60)
print(f"{'Trung bình cộng': <15} {mean_data[0] :>10,.0f} {mean_data[1]:>13,.0f} {mean_data[2]:>13,.0f}")
print(f"{'Trung vị': <16} {median_data[0] :>10,.0f} {median_data[1]:>13,.0f} {median_data[2]:>13,.0f}")
print(f"{'Độ lệch chuẩn': <18} {std_data[0] :>10,.0f} {std_data[1]:>13,.0f} {std_data[2]:>13,.0f}")
print(f"{'Giá trị bé nhất': <16} {min_data[0] :>10,.0f} {min_data[1]:>13,.0f} {min_data[2]:>13,.0f}")
print(f"{'Giá trị lớn nhất': <16} {max_data[0] :>9,.0f} {max_data[1]:>13,.0f} {max_data[2]:>13,.0f}")
print(f"{'Giá trị nan': <18} {sum_nan[0] :>9,.0f} {sum_nan[1]:>13,.0f} {sum_nan[2]:>13,.0f}")
print(f"{'So sánh mean vs median': <1} {dif_mean_median[0] :>3,.0f} {dif_mean_median[1]:>13,.0f} {dif_mean_median[2]:>13,.0f}")
print(f"{'Số lượng outlier': <15} {so_outlier[0] :>9,.0f} {so_outlier[1]:>13,.0f} {so_outlier[2]:>13,.0f}")
```
